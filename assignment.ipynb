````python
# ## 문제 2: 사칙연산 함수

# 두 숫자와 연산자를 입력받아 계산 결과를 반환(`return`)하는 함수를 작성하세요.

# ### 요구사항
# - 함수 이름: `calculate`
# - 매개변수: `a`, `b`, `operator`
# - 지원 연산자: `+`, `-`, `*`, `/`
# - 0으로 나눌 경우 `"0으로 나눌 수 없습니다"` 반환
# - 잘못된 연산자 입력 시 `"잘못된 연산자입니다"` 반환

# ### 실행 예시
# ```python
# print(calculate(10, 3, "+"))   # 13
# print(calculate(10, 3, "-"))   # 7
# print(calculate(10, 3, "*"))   # 30
# print(calculate(10, 3, "/"))   # 3.3333333333333335
# print(calculate(10, 0, "/"))   # 0으로 나눌 수 없습니다
# print(calculate(10, 3, "%"))   # 잘못된 연산자입니다

# """

def calculate(a, b, operator):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0:
            return "0으로 나눌 수 없습니다"
        return a / b
    else:
        return "잘못된 연산자입니다"
    
print(calculate(10, 3, "+"))
print(calculate(10, 3, "-"))
print(calculate(10, 3, "*"))
print(calculate(10, 3, "/"))
print(calculate(10, 0, "/"))
print(calculate(10, 3, "%"))
````

````python
# # 문제 3: 반환값이 여러 개인 함수

# 숫자 2개를 입력받아 합계와 평균을 동시에 반환하는 함수를 작성하세요.

# ### 요구사항
# - 함수 이름: `sum_and_avg`
# - 매개변수: `a`, `b`
# - 반환값: `(합계, 평균)` 튜플 형태

# ### 실행 예시
# ```python
# result = sum_and_avg(80, 90)
# print(result)          # (170, 85.0)

# s, a = sum_and_avg(75, 95)
# print(f"합계: {s}")    # 합계: 170
# print(f"평균: {a}")    # 평균: 85.0

# """

def sum_and_avg(a, b):
    total = a + b
    avg = total / 2
    return (total, avg)

result = sum_and_avg(80, 90)
print(result)

s, a = sum_and_avg(75, 95)
print(f"합계: {s}")
print(f"평균: {a}")
````

````python
## 문제 5: `*args` 활용 — 과목 점수 평균

# 여러 과목의 점수를 **개수 제한 없이** 입력받아 평균을 반환하는 함수를 작성하세요.

# ### 요구사항
# - 함수 이름: `get_average`
# - `*args`를 사용하여 가변 인수 처리
# - 입력값이 없으면 `0` 반환
# - 평균은 소수점 둘째 자리까지 반올림

# ### 실행 예시
# ```python
# print(get_average(80, 90, 100))        # 90.0
# print(get_average(75, 82, 91, 68, 95)) # 82.2
# print(get_average(100))                # 100.0
# print(get_average())                   # 0

# """

def get_average(*args):
    if len(args) == 0:
        return 0
    return round(sum(args) / len(args), 2)

print(get_average(80, 90, 100))
print(get_average(75, 82, 91, 68, 95))
print(get_average(100))
print(get_average())
````

````python
## 문제 6: 구구단 출력 함수

# 구구단을 출력하는 함수를 작성하세요.

# ### 요구사항
# - 함수 이름: `print_gugudan`
# - 매개변수: `dan`, `start=1`, `end=9`
# - `for`문으로 구구단 출력
# - 잘못된 범위(`start > end`) 입력 시 `"잘못된 범위입니다"` 출력
# - 반환값: 없음 (`print()`로 직접 출력)

# ### 실행 예시
# ```python
# print_gugudan(5)
# # 5 x 1 = 5
# # 5 x 2 = 10
# # ...
# # 5 x 9 = 45

# print_gugudan(3, 3, 7)
# # 3 x 3 = 9
# # 3 x 4 = 12
# # 3 x 5 = 15
# # 3 x 6 = 18
# # 3 x 7 = 21

# print_gugudan(7, 9, 2)
# # 잘못된 범위입니다

def print_gugudan(dan, start=1, end=9):
    if start > end:
        print("잘못된 범위입니다")
        return
    for i in range(start, end + 1):
        print(f"{dan} x {i} = {dan * i}")

print_gugudan(5)
print()

print_gugudan(3, 3, 7)
print()

print_gugudan(7, 9, 2)
````

````python
## 문제 10: 큰 수 찾기 함수

# 숫자 3개를 입력받아 가장 큰 수를 반환하는 함수를 작성하세요.

# ### 요구사항
# - 함수 이름: `find_max`
# - 매개변수: `a`, `b`, `c`
# - `max()` 함수를 사용하지 말고 직접 비교해서 구현

# ### 실행 예시
# ```python
# print(find_max(3, 8, 5))    # 8
# print(find_max(10, 4, 7))   # 10

# """

def find_max(a, b, c):
    result = a
    if b > result:
        result = b
    if c > result:
        result = c
    return result

print(find_max(3, 8, 5))
print(find_max(10, 4, 7))
````

````python
# # 문제 13: 팩토리얼 함수

# 정수 `n`을 입력받아 `n!` 값을 반환하는 함수를 작성하세요.

# ### 요구사항
# - 함수 이름: `factorial`
# - 매개변수: `n`
# - `for`문 사용
# - `0! = 1` 처리

# ### 실행 예시
# ```python
# print(factorial(5))  # 120
# print(factorial(0))  # 1

# """

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
print(factorial(0))

````

````python
# ## 문제 16: 문자열 길이 구하기 함수

# 문자열을 입력받아 글자 수를 반환하는 함수를 작성하세요.

# ### 요구사항
# - 함수 이름: `my_len`
# - 매개변수: `text`
# - `len()` 함수 사용 없이 직접 구현

# ### 실행 예시
# ```python
# print(my_len("python"))   # 6
# print(my_len("안녕하세요")) # 5

# """

def my_len(text):
    count = 0
    for _ in text:
        count += 1
    return count

print(my_len("python"))
print(my_len("안녕하세요"))
````

````python
# ## 문제 17: 특정 문자 개수 세기

# 문자열과 문자를 입력받아 해당 문자가 몇 번 나오는지 반환하는 함수를 작성하세요.

# ### 요구사항
# - 함수 이름: `count_char`
# - 매개변수: `text`, `ch`
# - `count()` 메서드 사용 없이 직접 구현

# ### 실행 예시
# ```python
# print(count_char("banana", "a"))  # 3
# print(count_char("hello", "l"))   # 2

# """

def count_char(text, ch):
    count = 0
    for c in text:
        if c == ch:
            count += 1
    return count

print(count_char("banana", "a"))
print(count_char("hello", "l"))
````

````python
# ## 문제 18: 거꾸로 문자열 반환

# 문자열을 입력받아 뒤집은 결과를 반환하는 함수를 작성하세요.

# ### 요구사항
# - 함수 이름: `reverse_text`
# - 매개변수: `text`
# - 슬라이싱 `[::-1]` 없이 구현해 보기

# ### 실행 예시
# ```python
# print(reverse_text("python"))  # nohtyp
# print(reverse_text("abc"))     # cba

# """

def reverse_text(text):
    result = ""
    for i in range(len(text) - 1, -1, -1):
        result += text[i]
    return result

print(reverse_text("python"))
print(reverse_text("abc"))
````

````python
class Product:
    def __init__(self, product_id, name, price, weight, stock):
        self.prcodut_id = product_id
        self.name = name
        self.price = price
        self.weight = weight
        self.stock = stock
    
    def get_info(self):
        return {
            "product_id": self.product_id, 
            "name": self.name, 
            "price": self.price, 
            "weight": self.weight, 
            "stock": self.stock
            }
    
    def update_stock(self, quantity):
        if self.stock >= quantity:
            self.stock -= quantity
            return True
        else:
            return False
    
    def apply_discount(self, discount_rate):
        discounted_price = self.price * (1 - discount_rate)
        return int(discounted_price)
````

````python
class ShoppingCart:
    def __init__(self):
        self.items = []
    
    def add_product(self, product, quantity):
        if product.update_stock(quantity):
            self.items.append({'product': product, 'quantity': quantity})
            return f"[알림] {product.name} {quantity}개가 장바구니에 추가되었습니다."
        else:
            return f"[경고] {product.name}의 재고가 부족합니다. (현재 재고: {product.stock}개)"
    
    def remove_product(self, product_id):
        for item in self.items:
            if item['product'].product_id == product_id:
                item['product'].stock += item['quantity']
                self.items.remove(item)
                return f"[알림] 상품 ID {product_id}가 장바구니에서 제거되었습니다."
        return f"[경고] 해당 상품이 장바구니에 없습니다."
    
    def get_total_price(self):
        total = 0
        for item in self.items:
            total += item['product'].price * item['quantity']
        return total
    
    def get_total_weight(self):
        total = 0
        for item in self.items:
            total += item['product'].weight * item['quantity']
        return total
    
    def show_cart(self):
        result = "\n=== 장바구니 내역 ===\n"
        if not self.items:
            result += "장바구니가 비어 있습니다.\n"
        else:
            for item in self.items:
                p = item['product']
                q = item['quantity']
                result += f"- {p.name}: {q}개 (개당 {p.price:,}원)\n"
            result += f"총 금액: {self.get_total_price():,}원\n"
            result += f"총 무게: {self.get_total_weight():.1f}kg\n"
        result += "===================\n"
        return result
````

````python
class Order:  
    def __init__(self, order_id, cart, customer_name):
        self.order_id = order_id
        self.cart = cart
        self.customer_name = customer_name
        if self.cart.get_total_price() >= 50000:
            self.shipping_fee = 0
        else:
            self.shipping_fee = 3000
    
    def calculate_final_price(self):
        return self.cart.get_total_price() + self.shipping_fee
    
    def print_receipt(self):
        receipt = "=" * 30 + "\n"
        receipt += "          주 문 서          \n"
        receipt += "=" * 30 + "\n"
        receipt += f"주문 ID : {self.order_id}\n"
        receipt += f"고객명  : {self.customer_name}\n"
        receipt += "-" * 30 + "\n"
        receipt += "[주문 상품 목록]\n"
        for item in self.cart.items:
            p = item['product']
            q = item['quantity']
            receipt += f"{p.name} x {q} = {p.price * q:,}원\n"
        receipt += "-" * 30 + "\n"
        receipt += f"상품 금액 : {self.cart.get_total_price():,}원\n"
        receipt += f"배송비    : {self.shipping_fee:,}원\n"
        receipt += "-" * 30 + "\n"
        receipt += f"최종 결제 금액: {self.calculate_final_price():,}원\n"
        receipt += "=" * 30 + "\n"
        return receipt
````
