```python
# cal1 
# 10 더하고
# 2 빼고
# 5 곱하고
# 4 나눈값 

# cal2 
# 5 더하고 
# 10 곱하고
# 10 빼고
# 5 나눈값을
class Calculator:
    def __init__(self):
        self.result = 0 
    
    def add(self, num): 
        self.result = self.result + num 
        return self.result 
    
    def sub(self, num): 
        self.result -= num 
        return self.result 
    
    def mul(self, num): 
        self.result *= num 
        return self.result 
    
    def div(self, num): 
        self.result /= num 
        return self.result 
    

cal1 = Calculator()
cal1.add(10)
cal1.sub(2)
cal1.mul(5)
cal1.div(4)

cal2 = Calculator()
cal2.add(5)
cal2.mul(10)
cal2.sub(10)
cal2.div(5)
```

```python
class FourCal:
    def setdata(self, first, second): 
        self.first = first
        self.second = second 

    def add(self):
        result = self.first + self.second
        return result 
    def sub(self):
        result = self.first - self.second
        return result 
    def mul(self):
        result = self.first * self.second
        return result 
    def div(self):
        result = self.first / self.second
        return result 
```
